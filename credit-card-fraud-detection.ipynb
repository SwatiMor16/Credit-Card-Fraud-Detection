Importing the dependencies


```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading dataset into pandas dataframe 
credit_card_data = pd.read_csv('/kaggle/input/credit-card-dataset/creditcard.csv')
```

```python
# get the first 5 rows of a dataset 
credit_card_data.head()
```

```python
# last 5 rows of dataset

credit_card_data.tail()
```

```python
# Now we will get some information about the data set

credit_card_data.info()
```

```python
# other way of finding the number of missing values 

credit_card_data.isnull().sum()
```

```python
# checking the distribution of legit and fraud transactions 

credit_card_data['Class'].value_counts()
```

this data set is highly imbalanced dataset and we can not train our model with this data as it will always predict as legit transaction
so we will separate the data for analysis 

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measure of data 
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare both types of transactions 

credit_card_data.groupby('Class').mean()
```

so to resolve this issue we will use under sampling method - bulid a dataset containing
similar distribution of legit and fraudulent transactions

```python
legit_sample = legit.sample(n = 492)
```

```python
new_dataset = pd.concat([legit_sample , fraud] , axis = 0)
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# split data into fearures and targets and store them in variables X ,Y 

X = new_dataset.drop(columns = 'Class' , axis = 1)
Y = new_dataset['Class']
print(X)
print(Y)

```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size = 0.2,stratify = Y , random_state = 2)
print(X.shape , X_train.shape , X_test.shape)
```

```python
# Modle training - we will be using LogisticRegression Model as we have binary classification 

modle = LogisticRegression()
```

```python
modle.fit(X_train,Y_train)
```

```python
# evaluation of data based on accuracy score 

X_train_prediction = modle.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
print('Training data accuracy : ' , training_data_accuracy)
```

```python
X_test_prediction = modle.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction , Y_test)
print('Testing data accuracy : ' , test_data_accuracy)
```
